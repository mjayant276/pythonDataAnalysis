### Reading and working with datasets
- By convention, we import the module dask.dataframe as dd, and call the corresponding DataFrame object ddf.
- The term “Dask DataFrame” is slightly overloaded. Depending on the context, it can refer to the module or the DataFrame object
- The following filename includes a glob pattern *, so all files in the path matching that pattern will be read into the same DataFrame.

```python
import os
import dask.dataframe as dd

ddf = dd.read_csv(
    os.path.join("data", "stocks", "*.csv")
)
ddf
```

- Dask has not loaded the data yet, it has: - investigated the input path and found that there are ten matching files - intelligently created a set of jobs for each chunk – one per original CSV file in this case

#### Lazy Evaluation

- Most Dask Collections, including Dask DataFrame are evaluated lazily, which means Dask constructs the logic (called task graph) of your computation immediately but “evaluates” them only when necessary. You can view this task graph using .visualize().

```python
ddf.visualize()
```

#### Computations with dask.dataframe and pandas and compare the both

```python
import pandas as pd

files = os.listdir(os.path.join('C:\\Users\\Jayanth\\Python_DataAnalysis\\Dask_Lessons\\data', 'stocks'))

maxes = []

for file in files:
    df = pd.read_csv(os.path.join('C:\\Users\\Jayanth\\Python_DataAnalysis\\Dask_Lessons\\data', 'stocks', file))
    maxes.append(df.MARKETCAP.max())

final_max = max(maxes)
```

- <b>dask.dataframe</b> lets us write pandas-like code, that operates on larger-than-memory datasets in parallel.

```python
%%time
result = ddf.MARKETCAP.max()
result.compute()
```

### Solutions

```python
# 1. 
len(ddf)
```

```python
#2. 
ddf.groupby('ticker').ticker.count().compute()
```

```python
#3.
print(f"mean for adjusted close is {ddf.adj_close.mean().compute()}")
print(f"mean for high is {ddf.high.mean().compute()}")
print(f"mean for low is {ddf.low.mean().compute()}")
```

```python
#4. for each ticker the mean is
print(f"mean for ticker adjusted close is {ddf.groupby('ticker').adj_close.mean().compute()}")
print("=============")
print(f"mean for ticker high is {ddf.groupby('ticker').high.mean().compute()}")
print("=============")
print(f"mean for ticker low is {ddf.groupby('ticker').low.mean().compute()}")
```
