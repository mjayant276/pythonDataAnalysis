- Import the libraries needed for the dashboard to be opened

```python
import dask.dataframe as dd
from dask.distributed import Client
```

```python
client = Client()
client
```

- to use the below parquet from s2 need to install s3fs (pip install s3fs)

```python
ddf = dd.read_parquet(
    "s3://dask-data/nyc-taxi/nyc-2015.parquet/part.*.parquet",
    columns=["passenger_count", "tip_amount"],
    storage_options={"anon": True},
)
```

### The processing of the below cell can be seen on the dask dashboard

```python
result = ddf.groupby("passenger_count").tip_amount.mean().compute()
result
```
