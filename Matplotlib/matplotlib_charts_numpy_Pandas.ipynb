```python
import matplotlib.pyplot as plt

feedata = [300, 500, 700]
feelabels = 'admission fee', 'registration fee', 'course fee'
colors = ['lightblue', 'lightsteelblue', 'silver']
explode = (0, 0.08, 0)
plt.pie(feedata, labels=feelabels, autopct='%1.1f%%', startangle=15, shadow=True, colors=colors, explode=explode)
plt.title('Fees Distribution')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
   
data = {'state': ['A', 'B', 'C', 'D', 'E'],
        'population': [45000, 42000, 52000, 49000, 47000]
       }
df = pd.DataFrame(data)

colors = ['green','blue','red','gold','magenta']
plt.bar(df['state'], df['population'], color=colors)
plt.title('State Vs Population', fontsize=12)
plt.xlabel('State', fontsize=12)
plt.ylabel('Populations', fontsize=12)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'inflation_rate': [5.1, 6.8, 6.7, 5.7, 6.8, 6.5, 5.5, 5.3, 6.2, 6.2],
        'gas_price': [1520, 1620, 1535, 1235, 1415, 1450, 1455, 1460, 1455, 1465]
        }

df = pd.DataFrame(data)

plt.scatter(df['inflation_rate'], df['gas_price'], color='darkblue')
plt.title('Inflation Rate Vs Gas Price', fontsize=14)
plt.xlabel('Inflation Rate', fontsize=14)
plt.ylabel('Gas Price', fontsize=14)
plt.grid(True)
plt.show()
```
