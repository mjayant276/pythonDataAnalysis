* import matplotlib and numpy 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import math
x_coords = np.linspace(0,100, 1000)
y_coords = 2 +  2 * np.sin(2 * x_coords)
```

```python
# plot
fig, ax = plt.subplots()

ax.plot(x_coords, y_coords, linewidth=2.0)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

* **plot()** can be used for plotting values on both x and y axis

```python
plt.plot([1, 2, 3, 4,5], [5,4,3,2,1])
plt.show()
```

```python


x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Plot with different curves")

plt.legend()

plt.show()
```

* Using numpy arrays for matplotlib

```python
x = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(x, x, 'b--', x, x**2, 'gs', x, x**3, 'r^')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('')
plt.show()
```

The **Figure** serves as the highest-level container encompassing all plot elements. You can envision the Figure object as a box-like enclosure that houses one or more Axes.

**Axes**, on the other hand, symbolize individual plots. Within the **Axes** object, you'll find smaller components like axes, tick marks, lines, legends, titles, and text-boxes.

```python
%matplotlib inline
x1 = np.linspace(0, 15, 100)
fig, ax = plt.subplots(2)
# Call plot() method on the appropriate object
ax[0].plot(x1, np.sin(x1), 'g-')
ax[1].plot(x1, np.cos(x1), 'g-');
```

* Using the figure instance to draw the plot

```python
fig = plt.figure()

x2 = np.linspace(0, 5, 10)
y2 = x2 ** 2

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x2, y2, 'r')

axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('curve on the x and y ');
```

* Histograms are used to plot frequencies.

```python
data1 = np.random.randn(1000)

plt.hist(data1); 
```

```python


x1 = np.random.normal(0, 4, 1000)
x2 = np.random.normal(-2, 2, 1000)
x3 = np.random.normal(1, 5, 1000)

kwargs = dict(histtype='stepfilled', alpha = 0.5, density = True, bins = 40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)

plt.show();

```

* Multiple Bar Chart

```python
data3 = [[15., 25., 40., 30.],
         [11., 23., 51., 17.],
         [16., 22., 52., 19.]]

z1 = np.arange(4)

plt.bar(z1 + 0.00, data3[0], color = 'r', width = 0.25)
plt.bar(z1 + 0.25, data3[1], color = 'g', width = 0.25)
plt.bar(z1 + 0.50, data3[2], color = 'b', width = 0.25)

plt.show()
```

* Stacked Bar Chart

```python
X = [10., 20., 30., 12.]

Y = [15., 25., 50., 20.]

z2 = range(4)

plt.bar(z2, X, color = 'b')
plt.bar(z2, Y, color = 'g', bottom = X)

plt.show()
```

```python
plt.figure(figsize=(7,7))

X = [35, 25, 20, 20]

labels = ['Railways', 'BusTransport', 'Airplane', 'Carrier']

plt.pie(X, labels=labels);

plt.show()
```

```python
plt.figure(figsize=(7,7))

x11 = [30, 25, 20, 15, 10]

labels = ['Subject1', 'Subject2', 'Subject3', 'Subject4', 'Subject5']

explode = [0.2, 0.1, 0.1, 0.05, 0]

plt.pie(x11, labels=labels, explode=explode, autopct='%1.1f%%');

plt.show()
```
