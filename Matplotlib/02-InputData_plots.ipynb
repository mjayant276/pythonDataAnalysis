#### In this exampe we shall see how different plots can be built using dict, the numpy array and the pandas dataframe

- **Passing dictionary as input**

```python
## Using dict
import matplotlib.pyplot as plt
import numpy as np
```

```python
## Passing a dictionary to the subplots
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(100)}
data['b'] = data['a'] + 10 * np.random.randn(50)
#data['d'] = np.abs(data['d']) * 50

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained') # constrained_layout automatically adjusts subplots and decorations like legends and colorbars so that they fit in the figure window 
ax.scatter('a', 'b', c='c', marker='^',  data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

- **Passing numpy array as input**

```python
import matplotlib.pyplot as plt
import numpy as np

# Some example data to display
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
# subplots() without arguments returns a Figure and a single Axes.
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('First introductory plot')
```

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
