#### 1. Replace all odd numbers in array with -9

```python
import numpy as np
arr1=np.array([1,3,6,5,8,7,9,12,15,19,18,20,17])

for i in range(len(arr1)):
    if arr1[i]%2!=0:
        arr1[i]=-9
        
print(arr1)        
```

#### 4. How to stack arrays vertically

```python
a1=np.array([1,2,3])
a2=np.array([4,5,6])

vertical_stack=np.vstack((a1,a2))

print(vertical_stack)
```

#### 5. How to stack horizontally

```python
a1=np.array([1,2,3])
a2=np.array([4,5])

horizontal_stack=np.hstack((a1,a2))

print(horizontal_stack)
```

#### 6. Reverse a vector np 1d array

```python
arr6=np.arange(1,11)
print(arr6)

reverse_array=arr6[::-1]
print(reverse_array)
```

#### 7. Draw a line in a diagram from position (1, 3) to position (8, 10)

#### 8. Create a Bar plot using with students data of student name and score.

```python
student_name = np.array(['Shreya', 'Vinayak','Rishabh','Ajin'])
score= np.array([87,80,90,98])

plt.ylim([70,100])

plt.bar(student_name,score)

plt.xlabel("student name")
plt.ylabel("scores")

plt.show()
```
